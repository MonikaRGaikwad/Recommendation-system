# Recommendation

# Book Dataset

```python
#import the required libraries
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
#read the file
#Encoding='latin-1' is used when the csv file is decoded
Book_df = pd.read_csv("/content/book (1).csv",encoding="latin-1")
Book_df[0:5]
```

# Data Processing

```python
#Here we are using the pivot_table to make interchange of index,column and values
#reset
user_Book_df = Book_df.pivot_table(index='User.ID',
                                 columns='Book.Title',
                                 values='Book.Rating').reset_index(drop=True)
user_Book_df
```

```python
user_Book_df.index=Book_df['User.ID'].unique()
user_Book_df
```

```python
#imputing the NaN values with 0(zero) values
user_Book_df.fillna(0,inplace=True)
user_Book_df
```

# cosine simillarties 

```python
#Calculating the cosine similarties between the users
user_sim=1-pairwise_distances(user_Book_df.values,metric='cosine')
user_sim
```

```python
#Storing the result in DataFrame
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```

```python
#setting the index and column names with User.ID
user_sim_df.index=Book_df['User.ID'].unique()
user_sim_df.columns=Book_df['User.ID'].unique()
user_sim_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df.iloc[0:5,0:5]
```

```python
#looking for the most similar users
user_sim_df.idxmax(axis=1)[0:5]
```
